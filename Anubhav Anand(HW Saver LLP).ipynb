```python
#importing the datasets for nltk and cleaning the datasets
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
raw = []
for i in range(0, 34116):
  cell = re.sub('[^a-zA-Z]', ' ', Raw_skills['Technology_Skills'][i])
  cell = cell.lower()
  cell = cell.split()
  ps = PorterStemmer()
  all_stopwords = stopwords.words('english')
  all_stopwords.remove('not')
  cell = [ps.stem(word) for word in cell if not word in set(all_stopwords)]
  cell = ' '.join(cell)
  raw.append(cell)
```

```python
#creating the train data
train_set = []
for i in range(0, 979):
  cell = re.sub('[^a-zA-Z]', ' ', Technical_skills['Technology_Skills'][i])
  cell = cell.lower()
  cell = cell.split()
  ps = PorterStemmer()
  all_stopwords = stopwords.words('english')
  all_stopwords.remove('not')
  cell = [ps.stem(word) for word in cell if not word in set(all_stopwords)]
  cell = ' '.join(cell)
  train_set.append(cell)
```

```python
#importing the CountVectorizer from the sKlearn library,and countVector are used for tokenize the sentence
from sklearn.feature_extraction.text import CountVectorizer
count_vec = CountVectorizer(max_features = 1500)
X = count_vec.fit_transform(train_set).toarray()
X_test = count_vec.transform(raw).toarray()
y = np.ones((979,))
```

```python
#importing the gaussian naive bayes classifier
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(X, y)
```

```python
#predect the value
select = classifier.predict(X_test)
```

```python
Info=[]
```

```python
for i in range(len(X_test)):
    if select[i] == 1:
        Info.append(X_test[i])
```

```python
#transform the data into orginal form
Info = count_vec.inverse_transform(Info)
```

```python
type(Info)
```

```python
After_tokenize = []
```

```python
for i in Info:
    if list(i) != []:
        After_tokenize.append(list(i))
```
